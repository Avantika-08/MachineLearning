# Seaborn
## Distribution plots
-> distplot
-> joinplot 
-> pairplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

## Correlation with heatmap
Heatmap is helps to find the correlation between each and every feature

```python
df.dtypes
```

```python
df_number = df.select_dtypes(include='number')
df_number.corr()
```

```python
sns.heatmap(df_number.corr())
```

## Join Plot
it helps to do bivariate analysis(2 variables)

```python
sns.jointplot(x='tip',y='total_bill',data=df_number,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df_number,kind='reg')
```

## Pair Plot
when we have more than 2 independent feature. Also known as pairplot. Value needs to be integer or floating value

```python
sns.pairplot(df_number)
```
