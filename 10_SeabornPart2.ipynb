```python
import seaborn as sns
```

## Categorical Plots
->Boxplot 
->Violinplot
->Countplot
->Barplot 

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.countplot(data=df, x='day')
```

```python
sns.countplot(data=df, y='sex')
```

```python
# Barplot
sns.barplot(y='total_bill',x='sex',data=df)
```

# Boxplot

```python
sns.boxplot(data=df, x="smoker", y="total_bill")
```

```python
sns.boxplot(data=df, x="day", y="total_bill",palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorising data based on some other categories
sns.boxplot(data=df,x='total_bill',y='day',hue='smoker')
```

# Violinplot

```python
sns.violinplot(data=df,x='total_bill',y='day',palette='rainbow')
```
