## Matplotlib
base library -> Numpy

```python
import matplotlib.pyplot as plt

#using this we need not to write show it will directly plot the graph
%matplotlib inline       
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
```

```python
# Scatter

plt.scatter(x,y,c='r')           # c->color
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Graph")
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
# Plot
#plt.plot(x,y,'g-')              # -- is for multiple dashed lines
#plt.plot(x,y,'g*-')             # *- lines
#plt.plot(x,y,'go')               # dotted lines
#plt.plot(x,y,'go--')
plt.plot(x,y,'r*',linestyle='dashed',linewidth=1,markersize=12)  
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plot Graph")
plt.savefig('Test1.png')
```

```python
# Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'ro')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(x,y,'c')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib Demo")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y)
```

```python
# Sine Curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("Sine wave form")

plt.plot(x,y)
plt.show()
```

```python
# Subplots
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)

plt.plot(x,y_sin,'r')
plt.title("Sine Wave")

plt.subplot(2,1,2)

plt.plot(x,y_cos,'b')
plt.title("Cosine Wave")

plt.savefig('SineCosine.png')
```
