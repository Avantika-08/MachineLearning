## Matplotlib
base library -> Numpy

```python
import matplotlib.pyplot as plt

#using this we need not to write show it will directly plot the graph
%matplotlib inline       
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
```

```python
# Scatter

plt.scatter(x,y,c='r')           # c->color
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Graph")
```
